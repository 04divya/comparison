DOCUMENT 1

```python
from google.colab import files

uploaded = files.upload()

!apt-get install poppler-utils

from pdf2image import convert_from_path
import pytesseract
import os
import json
import cv2
import numpy as np
from PIL import Image
```

```python
pdf_doc1 = list(uploaded.keys())[0]

pages_doc1 = convert_from_path(pdf_doc1, 400)

def preprocess_image(image):
    image = np.array(image)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  # Convert to Grayscale
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)  # Binarization
    denoised = cv2.medianBlur(binary, 3)  # Noise removal
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2,2))
    sharpened = cv2.morphologyEx(denoised, cv2.MORPH_GRADIENT, kernel)  # Sharpen edges
    return Image.fromarray(sharpened)  # Convert back to PIL format

text_doc1 = ""

for page in pages_doc1:
    processed_page = preprocess_image(page)  # Apply preprocessing
    text_doc1 += pytesseract.image_to_string(processed_page)  # OCR extraction

output = {"extracted_text": text_doc1}

with open("output1.json", "w") as f:
    json.dump(output, f)

print(text_doc1)
```

DOCUMENT 2

```python
# Step 1: Upload & Import Document 2
from google.colab import files

uploaded = files.upload()

# Step 3: Import required libraries
from pdf2image import convert_from_path
import pytesseract
import os
import json
import cv2
import numpy as np
from PIL import Image

pdf_doc2 = list(uploaded.keys())[0]

pages_doc2 = convert_from_path(pdf_doc2, 400)

def preprocess_image(image):
    image = np.array(image)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  # Convert to Grayscale
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)  # Binarization
    denoised = cv2.medianBlur(binary, 3)  # Noise removal
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2,2))
    sharpened = cv2.morphologyEx(denoised, cv2.MORPH_GRADIENT, kernel)  # Sharpen edges
    return Image.fromarray(sharpened)  # Convert back to PIL format

text_doc2 = ""

for page in pages_doc2:
    processed_page = preprocess_image(page)  # Apply preprocessing
    text_doc2 += pytesseract.image_to_string(processed_page)  # OCR extraction

output = {"extracted_text": text_doc2}

with open("output2.json", "w") as f:
    json.dump(output, f)

print(text_doc2)
```

FINDING SIMILARITIES

```python
# BERT-based Cosine Similarity

!pip install sentence-transformers

from sentence_transformers import SentenceTransformer, util
import json

# Load text from JSON files
with open("output1.json", "r") as f:
    data1 = json.load(f)
document1_text = data1["extracted_text"]

with open("output2.json", "r") as f:
    data2 = json.load(f)
document2_text = data2["extracted_text"]

model = SentenceTransformer('all-mpnet-base-v2')

# Compute sentence embeddings
embeddings1 = model.encode(document1_text)
embeddings2 = model.encode(document2_text)

# Compute cosine similarity
cosine_similarity = util.cos_sim(embeddings1, embeddings2)

# Convert to percentage
similarity_percentage = cosine_similarity.item() * 100

print(f"Cosine Similarity between documents: {similarity_percentage:.2f}%")
```

```python
# TF-IDF + Cosine Similarity
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

with open("output1.json", "r") as f:
    data1 = json.load(f)
with open("output2.json", "r") as f:
    data2 = json.load(f)

document1_text = data1["extracted_text"]  # Extract text content for Document 1
document2_text = data2["extracted_text"]  # Extract text content for Document 2

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([document1_text, document2_text])

#Calculate cosine similarity
similarity_score = cosine_similarity(vectors[0], vectors[1])[0][0]

similarity_percentage = round(similarity_score * 100, 2)

print(f"Cosine Similarity between documents: {similarity_percentage}%")
```
